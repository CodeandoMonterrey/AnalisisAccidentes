```python
import pandas as pd
import numpy as np
```

```python
import unicodedata
def strip_accents(s):
   return ''.join(c for c in unicodedata.normalize('NFD', s)
                  if unicodedata.category(c) != 'Mn')
```

```python
data = pd.read_csv("All_data_limpia.csv")
```

```python
data.columns
```

```python
data["Mes"] = data["Mes"].str.lower()
```

```python
data["Mes"].value_counts()
```

```python
data["Ciudad"] = data["Ciudad"].str.lower()
data["Ciudad"] = data["Ciudad"].str.strip()
data["Ciudad"] = data["Ciudad"].map(lambda x: strip_accents(str(x)))
data["Ciudad"].value_counts()
```

```python
for i in range(0,len(data["Fecha"]), 100):
    print(data["Fecha"][i])
    print(i)
```

```python
percances = []
for ciudad in data["Ciudad"].unique():
    ciudad_obj = {"Ciudad":ciudad}
    data_ciudad = data[data["Ciudad"] == ciudad]
    counts_ciudad = data_ciudad["Mes"].value_counts()
    for mes in counts_ciudad.keys():
        ciudad_obj[mes] = counts_ciudad[mes]
    percances.append(ciudad_obj)
```

```python
percances_df = pd.DataFrame(percances)
```

```python
data["Número total de muertos"] = pd.to_numeric(data["Número total de muertos"])
data["Número  total de lesionados"] = pd.to_numeric(data["Número  total de lesionados"])
```

```python
data["Número total de muertos"].sum()
```

```python
muertes = []
for ciudad in data["Ciudad"].unique():
    ciudad_obj = {"Ciudad":ciudad}
    data_ciudad = data[data["Ciudad"] == ciudad]
    counts_ciudad = data_ciudad["Mes"].value_counts()
    for mes in counts_ciudad.keys():
        ciudad_obj[mes] = data_ciudad[data_ciudad["Mes"]==mes]["Número total de muertos"].sum()
    muertes.append(ciudad_obj)
```

```python
lesionados = []
for ciudad in data["Ciudad"].unique():
    ciudad_obj = {"Ciudad":ciudad}
    data_ciudad = data[data["Ciudad"] == ciudad]
    counts_ciudad = data_ciudad["Mes"].value_counts()
    for mes in counts_ciudad.keys():
        ciudad_obj[mes] = data_ciudad[data_ciudad["Mes"]==mes]["Número  total de lesionados"].sum()
    lesionados.append(ciudad_obj)
```

```python
lesionados_df = pd.DataFrame(lesionados)
```

```python
muertes_df = pd.DataFrame(muertes)
```
